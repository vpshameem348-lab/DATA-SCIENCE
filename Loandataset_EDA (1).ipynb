## b, outlier detection and handling

```python
sns.histplot(num["past_due_days"],kde=True)
```

```python
x=["Principal","terms","past_due_days","age"]
for i in x:
    num.boxplot(column=[i])
    plt.title(f"Before Capping:{i}")
    plt.show()
```

```python
sns.boxplot(x="loan_status",y="terms",hue="Gender",data=num,palette="Set1")
```

```python
sns.boxplot(x="loan_status",y="terms",hue="Gender",data=num,palette="Set2")
```

### decide whether to cap, transform, or remove those outliers.
### campare the effect before and after capping by re-plotting Later.


```python
x=["Principal","terms","past_due_days","age"]
for i in x:
    q1 = num[i].quantile(0.25)
    q3 = num[i].quantile(0.75)
    iqr = q3 - q1
    lowercap = q1 - 1.5 * iqr
    uppercap = q3 + 1.5 * iqr
    num[i] = num[i].clip(lower=lowercap, upper=uppercap)
```

```python
# Values beyond 1.5 * IQR below Q1 or above Q3 are considered outliers.
# These outliers are capped to the nearest boundary (lower or upper).
```

```python
x=["Principal","terms","past_due_days","age"]
for i in x:
    plt.figure()
    num.boxplot(column=[i])
    plt.title(f"After Capping:{i}")
    plt.show()
```
